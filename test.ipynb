```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
weather = pd.read_csv('Test.csv')
weather.info()
```

```python
#sns.barplot(weather['humidity'], weather['temperature'])
#sns.barplot(x='humidity', y='temperature', data=weather)
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
ax = sns.barplot(x='humidity', y='temperature', data=weather)
ax.set_xticklabels(ax.get_xticks(), rotation=60)  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Ensures the plot layout is tight
plt.show()
```

```python
sns.displot(weather['humidity'])
```

```python
sns.displot(weather['humidity'], kde=False, rug=True);
```

```python
sns.jointplot(x='humidity', y='temperature', data=weather)
```

```python
sns.jointplot(x='humidity', y='temperature', data=weather, kind="hex")
```

```python
sns.jointplot(x='humidity', y='temperature', data=weather, kind="kde")
```

```python
sns.pairplot(weather[['humidity', 'temperature', 'air_pollution_index']])
```

```python
sns.stripplot(x='weather_type', y='temperature', data=weather)
```

```python
sns.stripplot(x='weather_type', y='temperature', data=weather, jitter=True)
```

```python
sns.swarmplot(x='humidity', y='temperature', data=weather)
```

```python
sns.boxplot(x='humidity', y='temperature', data=weather, hue=weather['weather_type'])
```

```python
sns.barplot(x='humidity', y='temperature', data=weather, hue=weather['weather_type'])
```

```python
sns.countplot(x='weather_type', data=weather)
```

```python
#sns.pointplot(x='humidity', y='temperature', data=weather, hue=weather['weather_type'])

plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
ax = sns.pointplot(x='humidity', y='temperature', data=weather, hue='weather_type')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Ensures the plot layout is tight
plt.show()
```
