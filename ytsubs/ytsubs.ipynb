```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
ytsubs = pd.read_csv('ytsubs.csv')
ytsubs.info()
```

```python
# Assuming your DataFrame is named 'df'
# Convert column 1 (Rank) to integers
ytsubs['Rank'] = pd.to_numeric(ytsubs['Rank'], errors='coerce', downcast='integer')

# Convert column 5 (Subscribers (millions)) to integers
ytsubs['Subscribers (millions)'] = pd.to_numeric(ytsubs['Subscribers (millions)'], errors='coerce', downcast='integer')
```

```python
sns.barplot(x='Rank', y='Subscribers (millions)', data=ytsubs)
```

```python
sns.displot(ytsubs['Subscribers (millions)'])
```

```python
sns.stripplot(x='Primary language', y='Subscribers (millions)', data=ytsubs)
plt.xticks(rotation=45)  # Adjust the rotation angle as needed
plt.show()
```

```python
sns.stripplot(x='Category', y='Subscribers (millions)', data=ytsubs)
plt.xticks(rotation=45)  # Adjust the rotation angle as needed
plt.show()
```

```python
sns.stripplot(x='Country', y='Subscribers (millions)', data=ytsubs)
plt.xticks(rotation=45)  # Adjust the rotation angle as needed
plt.show()
```

```python
#sns.boxplot(x='Subscribers (millions)', y='Rank', data=ytsubs, hue=ytsubs['Country'])

plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
sns.barplot(x='Rank', y='Subscribers (millions)', data=ytsubs, hue='Country')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Ensures the plot layout is tight
plt.show()
```
